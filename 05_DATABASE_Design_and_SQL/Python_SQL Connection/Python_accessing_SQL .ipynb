#### List all items of playlist, order by name descending

```python
%sql select * from playlists order by name desc
```

```python
%sql select * from tracks order by name desc limit 10
```

```python
%%sql
select GenreId, count(AlbumId) as 'Number of Albums by Genre'
from tracks
group by genreid
order by 2 desc
```

```python
%%sql
select distinct GenreId, count(AlbumId) over (partition by GenreId) as 'Number of Albums by Genre'
from tracks
order by 2 desc
```
