```python
import pandas as pd
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
import nltk
nltk.download("punkt")
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download("averaged_perceptron_tagger")
```

```python
f = open("data.txt", "r")
text = f.read()
```

```python
text[:1000]
```

```python
sent_tokens = sent_tokenize(text)
```

```python
len(sent_tokens)  # cümleler
```

```python
df = pd.DataFrame({"sentences": sent_tokens})  # herbir cümle bir row
```

```python
df.head()
```

```python
df.shape
```

```python
stop_words = stopwords.words('english')
lem = WordNetLemmatizer()
```

```python
def cleaning(data):
    #1. Tokenize
    text_tokens = word_tokenize(data)  # remove the .lower() in order to keep NNps, özel isimle isimiz yoksa .lower() duracak
    
    #2. Remove Puncs
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    #3. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    #4. Lemma
    text_cleaned = [lem.lemmatize(t) for t in tokens_without_sw]
    
    #5. joining
    return " ".join(text_cleaned)
```

```python
df["sentences_2"] = df["sentences"].apply(cleaning)  # satir satir cleaning uygula
```

**PoST (Part of Speech)**

```python
df["sentences_3"] = df["sentences_2"].apply(lambda x: x.split())  # joinsiz hali
```

```python
df.head()
```

```python
df["sentences_4"] = df["sentences_3"].apply(lambda x: nltk.pos_tag(x))  #  tagging yaptik
```
