# corr and cov

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp=[93,84,82,78,98,70]

number_of_people=[13,10, 11, 8, 15, 9]
```

```python
np.cov(temp, number_of_people)
```

```python
np.corrcoef(temp, number_of_people)
```
