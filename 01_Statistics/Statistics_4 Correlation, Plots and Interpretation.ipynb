## 1. Generate a scatter plot, calculate Pearson’s correlation, and interpret the relationship between two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0, 6, 4, 1, 5, 3, 4])
Y = np.array([88, 52, 64, 60, 54, 78, 40])
```

```python
plt.scatter(X, Y);  # SCATTER PLOT
```

```python
np.corrcoef(X,Y)[0,1]  # PEARSON' CORRELATION
```

### Interpretation: 

### In this situation there is a negative correlation and it means that as X variable increases, Y variable decreases. -0.673 correlation coefficient indicates that the correlation is very close to good correlation.

## 3. The last 10 game results between LA Lakers and LA Clippers are given in the table below.

```python
lakers = [103, 112, 106, 102, 87, 122, 105, 123, 107, 87]
```

```python
clippers = [101, 103, 111, 112, 93, 117, 113, 120, 118, 103]
```

```python
plt.boxplot(lakers, vert = False, whis = "range");
```

```python
plt.boxplot(clippers, vert = False, whis = "range");
```

### 1. Which team has a greater median, Lakers or Clippers?-----> Clippers
### 2. Which team has greater IQR, Lakers or Clippers? ------>  Clippers
### 3. Which team has a higher percentage of scores above its median? -----> Lakers

```python
from scipy.stats import iqr
import statistics as stat
q75, q25 = np.percentile(lakers, [75, 25])
print("median", stat.median(lakers))
print("q1:", q25)
print("q3:", q75)
print("iqr:", iqr(lakers))
lower_bound = q25-(1.5*iqr(lakers))
upper_bound = q75+(1.5*iqr(lakers))
print("lower bound:", lower_bound)
print("upper bound:", upper_bound)
print("min_value:", min(lakers))
print("max_vale:", max(lakers))
plt.boxplot(lakers, vert = False, whis = "range");
```

```python
from scipy.stats import iqr
import statistics as stat
q75, q25 = np.percentile(clippers, [75, 25])
print("median", stat.median(clippers))
print("q1:", q25)
print("q3:", q75)
print("iqr:", iqr(clippers))
lower_bound = q25-(1.5*iqr(clippers))
upper_bound = q75+(1.5*iqr(clippers))
print("lower bound:", lower_bound)
print("upper bound:", upper_bound)
print("min_value:", min(clippers))
print("max_vale:", max(clippers))
plt.boxplot(clippers, vert = False, whis = "range");
```
