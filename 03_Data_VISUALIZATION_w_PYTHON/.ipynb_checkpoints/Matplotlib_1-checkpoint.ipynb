```python
import matplotlib.pyplot as plt
```

**Functional Method**

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age,salary);  # or plot.show()  ## functional Method
```

```python
plt.plot(age,salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary vs Age");
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age,salary)
plt.plot(age, salary_2)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary vs Age");
```

```python
# one axes:
plt.plot(age,salary, label="salary")  # egriye label ile isimlendirme yapiyoruz
plt.plot(age, salary_2, label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary vs Age")
plt.legend();  # labellari gsteriyor sol stte
```

```python
# functional subplot:
plt.subplot(2, 1, 1)  # 2 satir 1 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age,salary, "r")
plt.subplot(2, 1, 2)
plt.plot(age, salary_2, "b");
```

```python
# functional subplot:
plt.subplot(1, 2, 1)  # 1 satir 2 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age,salary, "r")
plt.subplot(1, 2, 2)  # 1 satir 2 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age, salary_2, "b");
```

```python
# 4er plot
plt.subplot(2, 2, 1)  # 2 satir 2 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age,salary, "r")
plt.subplot(2, 2, 2)
plt.plot(age, salary_2, "b")
plt.subplot(2, 2, 3)  # 2 satir 2 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age,salary, "g")
plt.subplot(2, 2, 4)
plt.plot(age, salary_2, "y");
```

```python
# 6er plot
plt.subplot(2, 3, 1)  # 2 satir 2 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age,salary, "r")
plt.subplot(2, 3, 2)
plt.plot(age, salary_2, "b")
plt.subplot(2, 3, 3)  # 2 satir 2 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age,salary, "g")
plt.subplot(2, 3, 4)
plt.plot(age, salary_2, "y")
plt.subplot(2, 3, 5)  # 2 satir 2 sutun olacak. 3'nc rakam kacinci axes oldugunu gsteriyor
plt.plot(age,salary, "g")
plt.subplot(2, 3, 6)
plt.plot(age, salary_2, "y");
```

**OOP Method**

```python
fig = plt.figure()  # empty axes added
ax = fig.add_axes([0.0, 0.0, 0.4, 0.4])
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, 0.8, 0.8])
```

```python
# Manuel Design
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, 0.8, 0.8])

ax.plot(age, salary, "r");
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary vs Age");
```

```python
# Manual Design
fig = plt.figure()
ax1 = fig.add_axes([0.0, 0.0, 0.8, 0.8])
ax2 = fig.add_axes([1.0, 0.3, 0.4, 0.4])

ax1.plot(age, salary, "r");
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary vs Age");

ax2.plot(age, salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 vs Age");
```

```python
# Automatic Design : add_axes yapmadan direk geciyoruz
# one axes
fig, ax = plt.subplots(); # bire bire cizdirir default olarak, tek figure tek axes cizer
ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary vs Age");
```

```python
#2er axes
fig, ax = plt.subplots(2,1)

ax[0].plot(age, salary, "r")
ax[0].set_title("First Plot")

ax[1].plot(age, salary_2, "b")
ax[1].set_title("Second Plot")

plt.tight_layout()
```

```python
# 4er axes
fig, ax = plt.subplots(2,2)
ax[0,0].plot(age,salary, "r")
ax[0,0].set_title("salary by age")
ax[0,1].plot(age, salary_2)
ax[0,1].set_title("salary_2 by age")
ax[1,0].plot(age,salary, "g")
ax[1,0].set_title("salary by age")
ax[1,1].plot(age, salary_2, "y")
ax[1,1].set_title("salary_2 by age")
plt.tight_layout()
```

```python
fig, ax = plt.subplots(2,3)
ax[1,2].plot(age,salary)
plt.tight_layout();
```

```python
ig,ax=plt.subplots(nrows=3, ncols=2)
for i,j in range(3,2):
    ax[i,j].plot(age, salary, 'b')
    ax[i,j].set_xlabel('age')
    ax[i,j].set_ylabel('salary')
    ax[i,j].set_title('salary by age')
plt.tight_layout()
```
