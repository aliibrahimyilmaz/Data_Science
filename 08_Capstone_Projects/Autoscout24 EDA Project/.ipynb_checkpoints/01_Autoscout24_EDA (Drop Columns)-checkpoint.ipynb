### 1. Step : Drop columns that does not make any sense for the price of cars and that contain lots of NaN 

```python
df = df.drop("url", axis = 1)  # url col is dropped.
```

```python
df.shape
```

```python
df.isna().sum().sort_values(ascending = False)
```

```python
def show_nans(oto, limit):
    missing = oto.isnull().sum()*100/oto.shape[0]
    return missing.loc[lambda x : x > limit]
show_nans(df,35)  # %35 ten daha fazla NaN olanlari veriyor
```

```python
int(df.shape[0] * .40)  # % 40 and above non_NaNs survive 
```

```python
df1 = df.dropna(thresh=int(df.shape[0] * .40), axis=1)  # Only drop columns which have at least 40 % non-NaNs and %60 NaNs
# The parameter thresh=N requires that a column has at least N non-NaNs to survive.
# Think of this as the lower limit for missing data you will find acceptable in your columns.
```

```python
df1.isna().sum().sort_values(ascending = False)
```
