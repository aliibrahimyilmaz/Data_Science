## Calculate RFM Scoring

The simplest way to create customer segments from an RFM model is by using **Quartiles**. We will assign a score from 1 to 4 to each category (Recency, Frequency, and Monetary) with 4 being the highest/best value. The final RFM score is calculated by combining all RFM values. For Customer Segmentation, you will use the df_rfm data set resulting from the RFM analysis.

### i. Creating the RFM Segmentation Table


1. Create two functions, one for Recency and one for Frequency and Monetary. For Recency, customers in the first quarter should be scored as 4, this represents the highest Recency value. Conversely, for Frequency and Monetary, customers in the last quarter should be scored as 4, representing the highest Frequency and Monetary values.

```python
RFM_df.describe().T
```

```python
pd.qcut(RFM_df["Recency"], q=4).head()
```

```python
# first art of function
def recency_score(x):
    if x["Recency"] <= 19:
        recency = 4
    elif x["Recency"] <= 52:
        recency = 3
    elif x["Recency"] <= 144:
        recency = 2
    else:
        recency = 1
    return recency
```

```python
# second art of function
def frequency_score(x):
    if x <= 1:
        frequency = 1
    elif x <= 2:
        frequency = 2
    elif x<= 5:
        frequency = 3
    else:
        frequency = 4
    return frequency
```

**_Another simple way to calculate a Monetary score is to use tertiles. The ‘qcut’ function of pandas will divide the entire range of unique “Monetary” in 4 equal parts._**

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k = (2,10))
visu.fit(RFM_df1)
visu.poof()
```

### ii. Calculating time offset in Months i.e. Cohort Index:
Calculating time offset for each transaction will allows us to report the metrics for each cohort in a comparable fashion.
First, you will create 4 variables that capture the integer value of years, months for Invoice and Cohort Date using the get_date_int() function which you'll create it below.

```python
cohort_df.sample(10).InvoiceMonth.dt.year  # this gives a series of year
```

```python
def get_date_int(df, column):    
    year = df[column].dt.year    
    month = df[column].dt.month    
    day = df[column].dt.day
    return year, month, day
```

We will use this function to extract the integer values for Invoice as well as Cohort Date in 3 seperate series for each of the two columns

```python
invoice_year, invoice_month, _ = get_date_int(cohort_df, 'InvoiceMonth')  # results are pandas series
cohort_year, cohort_month, _ = get_date_int(cohort_df, 'CohortMonth')
```

```python
# results are also pandas series
years_diff = invoice_year - cohort_year  
months_diff = invoice_month - cohort_month
cohort_df['CohortIndex'] = years_diff * 12 + months_diff  # the diffrence 
cohort_df.head()
```

```python
cohort_df["TotalSales"] = cohort_df["Quantity"] * cohort_df["UnitPrice"]
cohort_df.head()
```

**_Now, we will obtain the cohort analysis matrix by grouping the data by CohortMonth and CohortIndex and aggregating on the TotalSales column by applying the pd.Series.sum() function. Here is the pivot cohort:_**

```python
third_df = cohort_df.groupby(["CohortMonth", "CohortIndex"], as_index= False)["TotalSales"].sum()
third_df
```

```python
pivoted3_df = third_df.pivot(index="CohortMonth", columns="CohortIndex", values="TotalSales")
pivoted3_df
```

```python
cohort_sizes = pivoted3_df.iloc[:,0]
retention3 = pivoted3_df.divide(cohort_sizes, axis=0)
```

```python
retention3.round(3) * 100
```

### ii. Visualize analysis of cohort 3 using seaborn and matplotlib modules

```python
# Drawing a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data = retention3, annot = True, fmt = '.0%', vmin = 0.0, vmax = 0.5, cmap = 'BuGn')
plt.title('Average Sales(Revenue) Retention rates')
plt.show()
```

```python
# Drawing a chart
plt.figure(figsize=(12, 8))
ax = sns.lineplot(x="CohortIndex", y="TotalSales", hue= "CohortMonth", data = third_df)
plt.title('Average Sales(Revenue) Retention rates')
ax.legend(loc='upper center', frameon=False)
plt.show()
```
